[UCI的Adult数据集](https://archive.ics.uci.edu/ml/datasets/Adult)

```python
import process
import lr
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score

train = process.process('data/adult.data', None)
test = process.process('data/adult.test', [0])

del train['native_country_Holand-Netherlands']

X_train = train.drop(['salary'], axis=1).as_matrix()
y_train = train.salary.as_matrix()

X_test = test.drop(['salary'],axis=1).as_matrix()
y_test = test.salary.as_matrix()
y_test = y_test.reshape(1, len(y_test))

clf = lr.LogisticRegression(train.drop(['salary'], axis=1).columns)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
```

```python
y_t = y_test.reshape((15055,))
y_p = y_pred.reshape((15055,))
```

```python
y_p[y_p >= 0.00000001] = 1
y_p[y_p < 0.00000001] = 0
```

```python
acc = accuracy_score(y_t, y_p)
print('accuracy:', acc)
```

```python
pd.DataFrame({'y_test':y_t, 'y_pred': y_p})
```
